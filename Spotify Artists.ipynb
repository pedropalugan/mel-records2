# Spotify Artists
### Importando bibliotecas e os dados

```python
import pandas as pd 
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
artistas_df = pd.read_csv('spotify_artists.csv')
display(artistas_df.head())
```

### Retirando colunas desnecessárias: unnamed, type

```python
artistas_df = artistas_df.drop('Unnamed: 0', axis=1)  
artistas_df = artistas_df.drop('type', axis=1)
display(artistas_df.head(20))
```

```python
artistas_df.describe()  # Descrição estatística dos dados
```

```python
artistas_df['followers'].max()
```

```python
artistas_df.sort_values(by='followers', ascending=False) #Comando para ordenar os artistas por seguidores
```

### Alguns gráficos
Histograma da popularidade

```python
plt.hist(artistas_df['artist_popularity'])
plt.title('Distribuição de Popularidade dos Artistas')
plt.xlabel('Popularidade') 
plt.ylabel('Frequência')
plt.show()
```

### Popularidade x Número de seguidores

```python
fig = px.scatter(artistas_df, x='followers', y='artist_popularity', title='Popularidade vs Seguidores dos Artistas', labels={'artist_popularity': 'Popularidade', 'followers': 'Número de Seguidores'})
fig.show()
```

Alguns artistas possuem muito mais seguidores que outros, mas a popularidade nao varia tanto

### Valores nulos

```python
artistas_df.isnull().sum()
```

```python
artistas_df = artistas_df.dropna()  # Remover linhas com valores ausentes
```

```python
(artistas_df['genres'] == '[]').sum() # Verifica se há gêneros vazios
```

Das 56128 linhas, 23540 estão sem gênero: por ser uma grande quantidade, não podemos descartá-las

```python
artistas_df['name'].duplicated().any()
```

```python
artistas_df['genres'] = artistas_df['genres'].replace('[]', np.nan)
```

```python
genero_por_nome = (artistas_df[artistas_df['genres'].notna()].drop_duplicates(subset='name').set_index('name')['genres'].to_dict())
artistas_df['genres'] = artistas_df.apply(lambda row: genero_por_nome.get(row['name'], row['genres']) if pd.isna(row['genres']) else row['genres'],axis=1)
display(artistas_df)
```

```python
print(artistas_df['genres'].isna().sum())
```

```python
artistas_df['genres'].value_counts()
```

```python
contagem_genero = artistas_df['genres'].value_counts().reset_index()
contagem_genero.columns = ['genres', 'quantidade']
fig = px.bar(contagem_genero, x='genres', y='quantidade', title='Quantidade de Gêneros Musicais', labels={'genres': 'Gênero Musical', 'quantidade': 'Quantidade'})
fig.show()
```
