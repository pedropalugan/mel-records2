```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
artistas_df = pd.read_pickle('artistas_df.pkl')
track_df = pd.read_pickle('spotify_tracks.pkl')
albums_df = pd.read_pickle('spotify_albums.pkl')
tabela_nova = pd.merge(artistas_df, albums_df, on='track_id', how='right')
display(tabela_nova)
```

para dar merge com a tabela de tracks, vamos usar o artists_id. Entretanto, a tabela tracks está com colchetes e aspas no ID.

```python
track_df['artists_id'] = track_df['artists_id'].str.replace(r"[\[\]']", '', regex=True)
```

Na outra tabela, falta um "s" em artists_id, então vamos mudar o nome da coluna

```python
tabela_nova.rename(columns={'artist_id': 'artists_id'}, inplace=True)
```

```python
tabela_nova2 = pd.merge(track_df, tabela_nova, on='artists_id', how='right')
display(tabela_nova2)
```

```python
tabela_nova2 = tabela_nova2.drop(columns=['available_markets_x', 'artists_id', 'available_markets_y', 'track_id'])
display(tabela_nova2)
```

### Completando os gêneros vazios
(pegar playlists iguais e completar as linhas dessas playlists que tenham generos vazios)


```python
tabela_nova2['genres'] = tabela_nova2.groupby('playlist')['genres'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

```python
print(tabela_nova2['genres'].isna().sum())
```

```python
tabela_nova2 = tabela_nova2.drop(['album_id', 'id_x', 'track_name_prev_x', 'id_y'], axis=1)
display(tabela_nova2)
```

```python
tabela_nova2.rename(columns={'name_x': 'name_artist'}, inplace=True)
tabela_nova2.rename(columns={'name_y': 'name_album'}, inplace=True)
tabela_nova2 = tabela_nova2.drop_duplicates()
```

```python
import ast
tabela_nova2['genres'] = tabela_nova2['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
generos_explodidos = tabela_nova2.explode('genres')
generos_explodidos['genres'] = generos_explodidos['genres'].astype(str).str.strip()
generos_explodidos = generos_explodidos.dropna(subset=['genres'])
generos_explodidos = generos_explodidos[generos_explodidos['genres'] != '']
top_generos = generos_explodidos['genres'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_generos.plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Gêneros Mais Frequentes', fontsize=16)
plt.xlabel('Gênero', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
